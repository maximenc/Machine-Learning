# **2. Datascience Libraries**

The `import` statement is used to import external libraries or modules, allowing you to use their functions and variables in your program.

ex:


```python
import numpy as np
data = [1, 2, 3, 4, 5]
print(f"Mean of the data: {np.mean(data)}")
```

Here, the `import numpy as np` statement imports the NumPy library and gives it the alias `np`, making it easier to use. In this example, the `mean` function from NumPy is accessed using `np.mean()`.

#### **2.1 NumPy**


**Creating Arrays**: using `np.array()`:


```python
# Create a 1D array
data = np.array([1, 2, 3, 4, 5])
print(data)
```

**multi-dimensional arrays:**

```python
# Create a 2D array (matrix)
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(matrix)
```

**Array Operations**:

```python
# Element-wise addition
data1 = np.array([1, 2, 3])
data2 = np.array([4, 5, 6])
sum_array = data1 + data2
print(sum_array)
```

**mean, sum:**

```python
# Calculate the mean and sum
mean_value = np.mean(data)
sum_value = np.sum(data)
print(f"Mean: {mean_value}, Sum: {sum_value}")
```

**Matrix multiplication**

```python
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])
product = np.dot(matrix1, matrix2)
print(product)
```

**Data Simulation:**

```python
# Generate random numbers
random_data = np.random.rand(5)
print(random_data)
```

```python
# Generate a random sample of 1000 numbers from a normal distribution
data = np.random.normal(loc=50, scale=10, size=1000)

# Calculate statistical measures
mean_value = np.mean(data)
median_value = np.median(data)
variance_value = np.var(data)

print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Variance: {variance_value}")

# Random sampling: choose 10 random elements from the data
sample = np.random.choice(data, 10, replace=False)
print("Random Sample:", sample)
```

You can also access rows by index using the `loc` and `iloc` methods:

```python
# Access rows using loc (by label) or iloc (by index)
row = df.loc[0]   # Access the first row by label
row = df.iloc[0]  # Access the first row by index
print(row)
```

**Filtering Data:**

Pandas allows you to filter rows based on conditions, which is very useful for exploring or preprocessing datasets:

```python
# Filter rows where 'Age' is greater than 30
filtered_df = df[df['Age'] > 30]
print(filtered_df)
```

**Data Manipulation**:

```python
# Add a new column 'Net Salary' by subtracting tax from 'Salary'
df['Net Salary'] = df['Salary'] - df['Salary'] * 0.2
print(df)
```

**Handling Missing Data:**

```python
# Drop rows with missing values
df_cleaned = df.dropna()

# Fill missing values in a column with the mean of that column
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

**Grouping and Aggregating:**

You can use the `groupby()` function to group data and calculate aggregated statistics like sum, mean, etc.:

```python
# Group by 'Age' and calculate the average salary for each group
grouped_df = df.groupby('Age')['Salary'].mean()
print(grouped_df)
```

**Merging and Joining Data:**

```python
# Merge two DataFrames on a common column
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [1, 2, 3], 'Score': [85, 90, 95]})
merged_df = pd.merge(df1, df2, on='ID')
print(merged_df)
```

#### **2.3 Matplotlib**

**Example 1: Simple Line Plot**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating data
x = np.linspace(0, 10, 100)  # 100 points evenly spaced between 0 and 10
y = np.sin(x)

# Adjusting figure size
plt.figure(figsize=(5, 3))

# Plotting the data
plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Sine of X')
plt.title('Sine Wave Plot')
plt.show()
```

**Example 2: Scatter Plot**

```python
# Generating random data for scatter plot
x = np.random.rand(50)
y = np.random.rand(50)

# Adjusting figure size
plt.figure(figsize=(5, 3))

# Plotting scatter plot
plt.scatter(x, y, edgecolor='black', color='red')
plt.xlabel('Random X')
plt.ylabel('Random Y')
plt.title('Scatter Plot')
plt.show()
```

**Example 3: Histogram**

```python
# Generating random data for histogram
data = np.random.randn(1000)  # 1000 normally distributed data points

# Adjusting figure size
plt.figure(figsize=(5, 3))

# Plotting histogram
plt.hist(data, bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

**Example 4: Bar Plot**

```python
# Defining categorical data
categories = ['A', 'B', 'C', 'D', 'E', 'F']
values = [10, 24, 36, 18, 5, 20]

# Adjusting figure size
plt.figure(figsize=(5, 3))

# Plotting bar plot
plt.bar(categories, values, color='green', edgecolor='black', alpha=0.5)
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

**Example 5: Two Simulated Trajectories in Subplots**

```python
n_steps = 1000  # Number of steps in each trajectory

# Generating random steps (+1 or -1) for two random walks
steps_1 = np.random.choice([-1, 1], size=n_steps)
steps_2 = np.random.choice([-1, 1], size=n_steps)

# Simulating the trajectories by taking cumulative sum of steps
trajectory_1 = np.cumsum(steps_1)
trajectory_2 = np.cumsum(steps_2)

# Generating time steps (x-axis)
time = np.arange(n_steps)

plt.figure(figsize=(5, 5))

# First subplot: First random walk
plt.subplot(2, 1, 1)
plt.plot(time, trajectory_1, color='blue')
plt.xlabel('Time Steps')
plt.ylabel('Position 1')
plt.title('Simulated Random Walk Trajectory 1')

# Second subplot: Second random walk
plt.subplot(2, 1, 2)
plt.plot(time, trajectory_2, color='green')
plt.xlabel('Time Steps')
plt.ylabel('Position 2')
plt.title('Simulated Random Walk Trajectory 2')
plt.tight_layout()
plt.show()
```

### **2.4 Seaborn**

advanced plot can be found here: https://seaborn.pydata.org/

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Here's how to load the 'titanic' dataset and visualize it.

```python
# Load the 'titanic' dataset, which contains data about Titanic passengers
titanic = sns.load_dataset("titanic")

# Display the first few rows of the dataset
print(titanic.head())
```

**Example 1: Scatter Plot for Age vs. Fare**

Seaborn’s scatterplot() function allows you to plot relationships between two variables. You can also use regplot() or lmplot() to add a regression line for better trend visibility.

```python
# Scatter plot showing relationship between age and fare
plt.figure(figsize=(5, 3))
sns.scatterplot(x="age", y="fare", data=titanic)
plt.show()
```

**Example 2: Heatmaps for Class and Survival**

Heatmaps can reveal the relationship between survival rates and passenger class. We'll create a pivot table and visualize it.

```python
# Create a pivot table showing survival by class and sex
titanic_pivot = titanic.pivot_table(index="class", columns="sex", values="survived", aggfunc="mean", observed=False)

# Plot the heatmap
plt.figure(figsize=(5, 3))
sns.heatmap(titanic_pivot, annot=True, cmap="Blues")
plt.show()
```

**Example 3: Categorical Plotting for Survival by Class**

```python
# Bar plot showing survival by passenger class
plt.figure(figsize=(5, 3))
sns.barplot(x="class", y="survived", data=titanic)
plt.show()
```

**Example 4: Pair Plot for Passenger Attributes**

```python
# Pair plot for the titanic dataset, focusing on numeric variables
sns.pairplot(titanic[['age', 'fare', 'pclass', 'survived']], hue="survived", height=1.55)  # Reduce plot size
plt.show()
```

**Example 5: Visualizing Data Distributions for Age**

```python
# Histogram for passenger age distribution
plt.figure(figsize=(5, 3))
sns.histplot(titanic['age'].dropna(), kde=True)
plt.show()
```
