# Scaling

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler, Normalizer, MaxAbsScaler

# Re-generate synthetic bivariate data with two clear clusters
np.random.seed(0)
n_datapoints = 50
x = np.random.normal(0, 1, size=(n_datapoints // 2, 2))
x = np.vstack([x, np.random.normal(4, 1, size=(n_datapoints // 2, 2))])

# Initialize scalers
scalers = {
    'Original Data': None,  # To compare with the original data
    'StandardScaler': StandardScaler(),
    'RobustScaler': RobustScaler(),
    'MinMaxScaler': MinMaxScaler(),
    'Normalizer': Normalizer(),
    'MaxAbsScaler': MaxAbsScaler(),
}

# Define the range for x and y axis
x_min, x_max = -3, 7
y_min, y_max = -3, 7

# Plot the data for each scaler
fig, axs = plt.subplots(2, 3, figsize=(15, 10))  # 2 lines, 3 plots each
axs = axs.flatten()  # Flatten to 1D array for easier iteration

for ax, (title, scaler) in zip(axs, scalers.items()):
    if scaler is not None:
        x_scaled = scaler.fit_transform(x)
    else:
        x_scaled = x
    ax.scatter(x_scaled[:, 0], x_scaled[:, 1], facecolors='grey', edgecolors='k', s=10)  # smaller circles
    ax.set_title(title)
    ax.set_xlim([x_min, x_max])
    ax.set_ylim([y_min, y_max])
    ax.grid(True)

plt.tight_layout()
plt.show()
```
