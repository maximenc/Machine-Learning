# Encoding

OneHotEncoder

```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Fetch the adult dataset
adult = fetch_openml(name='adult', version=1, as_frame=True)
df = adult.data

# Display the first few rows of the original 'workclass' data
print("Original 'workclass' data:")
print(df['workclass'].head())

# Selecting the 'workclass' column for one-hot encoding
workclass_data = df[['workclass']]

# Initialize OneHotEncoder
encoder = OneHotEncoder(sparse=False)

# Fit and transform the 'workclass' column
encoded_workclass = encoder.fit_transform(workclass_data)

# Convert the encoded data into a DataFrame
encoded_workclass_df = pd.DataFrame(encoded_workclass, columns=encoder.get_feature_names_out(['workclass']))

# Displaying the first few rows of the encoded DataFrame
print("\nOne-Hot Encoded 'workclass' data:")
print(encoded_workclass_df.head())
```
