### Sentiment Analysis with FinBERT

```python
############### FinBERT-Sentiment  ###############

import torch
from transformers import pipeline, BertTokenizer, BertForSequenceClassification

finbert_tone = BertForSequenceClassification.from_pretrained('yiyanghkust/finbert-tone',num_labels=3)
tokenizer_tone = BertTokenizer.from_pretrained('yiyanghkust/finbert-tone')
finbert_sent = pipeline("text-classification", model=finbert_tone, tokenizer=tokenizer_tone)

def get_sent_finbert(text):
    results = finbert_sent(text)
    class_ = results[0]["label"]
    return class_
```

```python
text = "environmental devastation"
get_sent_finbert(text)
```

### E, S, and G classification with FinBERT

```python
############### FinBERT-ESG  ###############

import torch
from transformers import pipeline, BertTokenizer, BertForSequenceClassification

finbert = BertForSequenceClassification.from_pretrained('yiyanghkust/finbert-esg',num_labels=4)
tokenizer = BertTokenizer.from_pretrained('yiyanghkust/finbert-esg')
finbert_esg = pipeline("text-classification", model=finbert, tokenizer=tokenizer)

def get_finbertesg(text):
    results = finbert_esg(text)
    label = results[0]["label"]
    return label
```

```python
text = 'environmental devastation'
get_finbertesg(text)
```
